```
#lendo o arquivo de dados
arquivo = "dbfs:/databricks-datasets/flights"
```

```
# lendo o arquivo de dados
# inferSchema = True (default true) se por False voce define o schema e passa como propriedade
# header = True
df = spark \
.read.format("csv")\
.option("inferSchema", "True")\
.option("header", "True")\
.csv(arquivo)
```

```
#imprime os datatypes das colunas do dataframe
df.printSchema()
```

```
df.show(2)
```

```
type(df)
```

```
# retorna as primeiras 5 linhas do dataframe em formato de array
df.take(5)
```

```
display(df.show(5))
```

```
df.count()
```

```
from pyspark.sql.functions import max
#posso usar o take ou show
df.select(max("delay")).show(1)
```

```
#filtrando linhas de um dataframe usando filter
df.filter("delay < 2").show(4)
```

```
#usando where (um alias para o metodo filter)
df.where("delay < 2").show(4)
```

```
from pyspark.sql.functions import desc, asc, expr
#ordenando por ordem crescente
df.orderBy(expr("delay desc")).show(10, truncate=False)
```

```
#visualizando estatisticas descritivas
df.describe().show()
```

```
# iterando sobre todas as linhas do dataframe
for i in df.collect():
  print(i)
  print(i[0], i[1] * 2)
```

```
# adicionado um anova coluna ao dataframe
df = df.withColumn('Nova Coluna', df['delay']+2)
df.show(10)
```

```
#removendo coluna
df = df.drop('Nova Coluna')
df.show(10)
```

```
#renomenado uma coluna no dataframe
df.withColumnRenamed('Nova Coluna', 'New Column').show()
```

```
df.filter('delay is NULL').show(10)
```

```
# outra forma filtrando valores missing
df.filter(df.delay.isNull()).show(10)
```

```
df.na.fill(value=0).show()
```

```
# preenche valores missing com valor 0 apenas da coluna que vc deseja
df.na.fill(value=0, subset=['delay']).show()
```

```
# preecnhe os dados com valores de string vazia
df.na.fill("").show()
```

```
# remover qualaquer lunha nula de qualquer coluna
df.na.drop().show()
```
